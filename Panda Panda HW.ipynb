```python
top_SN = new_df.groupby("SN").Price.agg(("count","mean", "sum"))
top_SN.sort_values('sum',ascending= False, inplace= True)

final_SN = top_SN.rename(columns={"count":"Purchase Count", "mean":"Average Purcase Price", "sum":"Total Purchase Value"})

final_SN.iloc[0:5]
```

```python
prof_item = merge_df.groupby(["Item Name"]).Price.agg(("count", "sum"),inplace=True)
prof_item.sort_values("sum", ascending= False, inplace= True)
prof_item["sum"] = '$' + round(prof_item['sum'],2).astype(str)

prof_item
```

```python
pop_item = merge_df.groupby(["Item Name"]).Price.agg(("count", "sum"),inplace=True)
pop_item.sort_values("count", ascending= False, inplace= True)
pop_item.head()
```
